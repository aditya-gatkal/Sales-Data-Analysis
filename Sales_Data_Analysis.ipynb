```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('sales_data.csv')
data.head()
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
print(data.isnull())
```

# Data Analysis (KPIs)

### Total Sales

```python
total_sales = data["Sales"].sum()
print(f"Total Sales: {total_sales}")
```

### Sales by Product

```python
sales_by_product = data.groupby('Product')['Sales'].sum()
print(f"Sales by product: {sales_by_product}")
```

### Sales by Region

```python
sales_by_region =data.groupby('Region')['Sales'].sum()
print(f"Sales by Region: {sales_by_region}")
```

### Monthly Sales Trend

```python
data["OrderDate"] = pd.to_datetime(data["OrderDate"])

data["Month"] = data["OrderDate"].dt.to_period("M")
monthly_sales = data.groupby('Month')['Sales'].sum()
print(f"Monthly Sales Trend: {monthly_sales}")
```




# Data Visualization

### Sales by Product (Bar Chart)

```python
sales_by_product.plot(kind ='bar',color ='skyblue',edgecolor ='black')
plt.title('Sales by Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

### Sales by Region (Pie Chart)

```python
sales_by_region.plot(kind='pie',autopct="%1.1f%%",startangle=90,shadow=True)
plt.title("Sales by Region")
plt.ylabel(" ")
plt.show()
```

### Monthly Sales Trend (Line Chart)

```python
monthly_sales.plot(kind='line',marker='o',color='red')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

### Quantity vs Sales (Scatter Plot)

```python
plt.scatter(data['Quantity'],data['Sales'],color='chocolate')
plt.title('Quantity vs Sales')
plt.xlabel('Quantity')
plt.ylabel('Sales')
plt.show()
```
